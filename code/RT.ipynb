```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
pd.options.display.float_format = '{:.2f}'.format
```

```python
df = pd.read_csv('../data/zippedData/rt.movie_info.tsv.gz' , delimiter='\t')
df.head()
```

```python
df.info()
```

```python
df['box_office'].describe()
```

```python
box_office_with_non_value = df.loc[df['box_office'].notnull()] 
box_office_with_non_value
```

```python
bownv = box_office_with_non_value.sort_values( by=['box_office'],ascending=False)
bownv
```

```python
bownv['box_office'] = bownv['box_office'].str.replace(',','')
```

```python
bownv['box_office'] = bownv['box_office'].astype(int) 
bownv['box_office']
```

```python
bownv.sort_values('box_office',ascending=False, inplace= True)
bownv
```

```python
bownv_top_100bo = bownv['rating'].head(100)
bownv_top_100bo.value_counts() # For the top 100 movies that have non-null box_office value, 
                               # looks like PG-13 appears the most.  
```

```python
bownv_top_100bo.value_counts().keys()
```

```python
fig,ax = plt.subplots()
x = bownv_top_100bo.value_counts().keys()
height = bownv_top_100bo.value_counts().values

ax.bar( x = x , height = height);
ax.set_xlabel('Content Rating')
ax.set_ylabel('# of Movies')
ax.set_title('Top 100 movies with the highest box office ');
```

```python
plot = sns.barplot(x = x , y = height, data = bownv_top_100bo , orient = 'v', color = '#00A4EF');
plot.set_xlabel('Content Rating')
plot.set_ylabel('# of Movies')
plot.set_title('Top 100 movies with the highest box office ');
```

```python
df_rating_avg_bf=bownv.groupby(by='rating').mean().reset_index()[['rating','box_office']]
df_rating_avg_bf.sort_values(by='box_office', ascending = False , inplace = True)
```

```python
df_rating_avg_bf.box_office = df_rating_avg_bf.box_office / 1000000 # make it simple
```

```python
fig, ax = plt.subplots()
ax.ticklabel_format(style='plain')
x = df_rating_avg_bf.rating
height = df_rating_avg_bf.box_office

ax.barh( y = x , width = height , color = 'red')
ax.set_xlabel('Average box office (Millions)')
ax.set_ylabel('Rating ')
ax.set_title('Average box office for each rating');
```

```python
x = df_rating_avg_bf.rating
height = df_rating_avg_bf.box_office
```

```python
df_rating_avg_bf.sort_values('box_office', ascending = True, inplace = True)
```

```python
plot = sns.barplot(x = height , y = x, data = df_rating_avg_bf , orient = 'h', color = '#00A4EF');
plot.set_xlabel('Revenve (Millions)')
plot.set_ylabel('Content Rating')
plot.set_title('Average Box Office Revenve By Content Rating');
plt.savefig("../Images/Average Box Office Revenve By Content Rating.jpg", dpi=400, bbox_inches='tight')
```

```python
df_rating_median_bf=bownv.groupby(by='rating').median().reset_index()[['rating','box_office']]
```

```python
df_rating_median_bf.sort_values(by='box_office', ascending = True , inplace = True)
```

```python
df_rating_median_bf.box_office = df_rating_median_bf.box_office / 1000000
```

```python
fig, ax = plt.subplots()
ax.ticklabel_format(style='plain')
plt.xticks(rotation= 60, ha="right")
x = df_rating_median_bf.rating
height = df_rating_median_bf.box_office

ax.barh( y = x , width = height);
ax.set_xlabel('Median box office (Millions)')
ax.set_ylabel('Rating ')
ax.set_title('Median box office for each rating');
```
